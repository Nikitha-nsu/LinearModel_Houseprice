## Without Scaling

```python
from sklearn.model_selection import train_test_split
X= df.drop(['id','price'],axis=1)
y = df.iloc[:,[1]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X.columns
```

```python
y.columns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
prediction = lm.predict(X_test)
```

```python
plt.scatter(y_test,prediction)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
```

```python
plt.plot(y_test,prediction)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
```

```python
coef = pd.DataFrame(data=lm.coef_ [0],index =X.columns, columns = ['Coefficient'])
coef
```

```python

sns.distplot(y_test-prediction,bins=50,axlabel= 'price')
```

```python
from sklearn import metrics
```

```python
#Linear model score
lm_score = lm.score(X_train,y_train)
lm_score
```

```python
print( 'Mean Absolute error', metrics.mean_absolute_error(y_test,prediction))
```

```python
print( 'Mean Squared error', metrics.mean_squared_error(y_test,prediction))
```

```python
print( 'R2 score', metrics.r2_score(y_test,prediction))
```

```python
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=10,random_state=42)
model.fit(X_train,y_train)
```

```python
#Estimate Decision tree model score
#model score
model_score = model.score(X_train,y_train)
model_score
```

```python
#Model prediction

dtree_pred = model.predict(X_test)
```

```python
#check r2score
dtree_r2 = metrics.r2_score(y_test,dtree_pred)
dtree_r2
```

```python
plt.scatter(y_test,dtree_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor(n_estimators=1000,random_state=42)
model2.fit(X_train,y_train)
```

```python
#Estimate Random Forest Regressor tree model score
#model score
model2_score = model2.score(X_train,y_train)
model2_score
```

```python
#model2 prediction
rregress_pred = model2.predict(X_test)
rregress_pred
```

```python
#model r2 score
model2_r2 = metrics.r2_score(y_test,rregress_pred)
model2_r2
```

```python
plt.scatter(y_test,rregress_pred)
```

## Cross Validation

```python
#Cross validation

from sklearn.cross_validation import cross_val_score, cross_val_predict
from sklearn import metrics# Perform 6-fold cross validation
```

```python
# Perform 6 fold cross validation
scores = cross_val_score(lm, X, y, cv=6)
print('Cross validated scores', scores)
```

```python
new_prediction = cross_val_predict(lm, X, y, cv=6)
plt.scatter(y,new_prediction)
```

```python
accuracy = metrics.r2_score(y,new_prediction)
print("Cross predicted accuracy", accuracy)
```
