## Train a model for house price prediction

```python
## Split the dataset into train and test set
```

```python
X= df.iloc[:,2:].values
y = df.iloc[:,[1]].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
## Create model and fit data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
prediction = lm.predict(X_test) #Prediction test set results
np.set_printoptions(precision=2) #Print values rounded upto two decimals
print(np.concatenate((prediction.reshape(len(prediction),1),y_test.reshape(len(y_test),1)),1)) 
# Comparing actual results and predicted results as two columns
```

```python
lm.coef_[0]
```

```python
# Verifying difference in results betweem Actual price and Predicted price
sns.distplot(y_test-prediction,bins=50,axlabel= 'price')
```

```python
from sklearn import metrics
```

```python
#Linear model score
lm_score = lm.score(X_train,y_train)
lm_score
```

```python
print( 'Mean Absolute error', metrics.mean_absolute_error(y_test,prediction))
```

```python
print( 'Mean Squared error', metrics.mean_squared_error(y_test,prediction))
```

```python
print( 'R2 score', metrics.r2_score(y_test,prediction))
```

```python
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=10,random_state=42)
model.fit(X_train,y_train)
```

```python
#Estimate Decision tree model score
#model score
model_score = model.score(X_train,y_train)
model_score
```

```python
#Model prediction
dtree_pred = model.predict(X_test)
```

```python
#check r2score
dtree_r2 = metrics.r2_score(y_test,dtree_pred)
dtree_r2
```

## Cross Validation

```python
#Cross validation

from sklearn.cross_validation import cross_val_score,cross_val_predict
from sklearn import metrics# Perform 5-fold cross validation
```

```python
# Perform 6 fold cross validation
scores = cross_val_score(lm, X, y, cv=5)
print('Cross validated scores', scores)
```

```python
new_prediction = cross_val_predict(lm, X, y, cv=5)
plt.scatter(y,new_prediction)
```

```python
accuracy = metrics.r2_score(y,new_prediction)
print("Cross predicted accuracy", accuracy)
```
